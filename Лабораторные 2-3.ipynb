```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import odeint

def dydt(x,t):
  k = 0.21
  return k*x
t = np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline 

plt.plot(t, solve, label = r"x'=k*x")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
url='https://raw.githubusercontent.com/ZhukovskayaT/MMVSULAB1/main/08.04.csv'
df=pd.read_csv(url)

years=np.array(df['year']).reshape(-1, 1)
population=np.array(df['population'])

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
year_poly = poly_features.fit_transform(years)
model = LinearRegression()
model.fit(year_poly, population)
predicted_population = model.predict(year_poly)
plt.plot(years, population)
plt.plot(years, predicted_population)
plt.plot(years, population)
plt.xlabel('year')
plt.ylabel('population')
plt.title('Численность населения г. Ростов-на-Дону')

plt.show()
     
```
