```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import odeint

def dydt(x,t):
  k = 0.21
  return k*x
t = np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline 

plt.plot(t, solve, label = r"x'=k*x")
plt.grid(True)
plt.legend()
plt.show()
```
